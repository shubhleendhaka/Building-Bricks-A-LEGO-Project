```python
import pandas as pd
```

```python
df = pd.read_csv('merged dataset.csv')
df.columns
```

```python
# Identify the top 6 theme_ids by frequency
top_6_theme_ids = df['theme_id'].value_counts().head(6).index
print(top_6_theme_ids)
```

```python
# Filter the DataFrame to include only rows with these top 6 theme_ids
subset_df = df[df['theme_id'].isin(top_6_theme_ids)]
```

```python
subset_df
```

```python
subset_df = subset_df.drop(columns=['Unnamed: 0'])
```

```python
subset_df.to_csv('top 6 themes subset.csv')
```
